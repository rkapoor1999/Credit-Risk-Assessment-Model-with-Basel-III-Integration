```python
# Initial exploration
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loans_df = pd.read_csv('../data/data.csv')
```

```python
print(loans_df.shape)
```

```python
print(loans_df.isnull().sum())
```

```python
# Visualize key distributions
plt.figure(figsize=(12, 6))
sns.histplot(loans_df['loan_amnt'])
plt.title('Distribution of Loan Amounts')
plt.show()
```

- The loan data shows a multimodal distribution with distinct spikes at standard loan amounts ($10K, $15K, $20K, $35K)
- Loans are predominantly concentrated in the $5,000-$25,000 range
- The non-normal distribution suggests lenders offer standardized loan packages rather than continuous amounts
- This pattern indicates potential risk segmentation based on loan size categories, which will be relevant for our Basel III risk weight assignments
- The discrete nature of loan amounts should be considered when developing our credit risk models

```python
# Look at relationships between key variables
sns.pairplot(loans_df[['loan_amnt', 'int_rate', 'annual_inc', 'dti']])
plt.show()
```
